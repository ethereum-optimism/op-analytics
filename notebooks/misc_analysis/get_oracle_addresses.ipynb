```python
## https://docs.chain.link/data-feeds/price-feeds/addresses?page=1&testnetPage=1

from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
results = []

for page in range(1, 17):
    url = f"https://docs.chain.link/data-feeds/price-feeds/addresses?page={page}&testnetPage=1&network=optimism"
    driver.get(url)
    time.sleep(2)

    # Only select rows from OP Mainnet section
    rows = driver.find_elements(By.XPATH, '//section[@id="op-mainnet"]//table//tbody/tr')
    for row in rows:
        try:
            address_elem = row.find_element(By.XPATH, './/a[contains(@href, "0x")]')
            address = address_elem.text.strip().lower()
            chain_id = 10
            project = 'chainlink'
            results.append((chain_id, address, project))
        except Exception:
            continue

driver.quit()

# Output in tuple format for SQL
for i, row in enumerate(results):
    tuple_str = f"({row[0]}, '{row[1]}', '{row[2]}')"
    if i < len(results) - 1:
        print(tuple_str + ',')
    else:
        print(tuple_str)

print("\nTotal addresses:", len(results))
```
