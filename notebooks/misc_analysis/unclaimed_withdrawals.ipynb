```python
import pandas as pd

df = pd.read_csv('inputs/withdrawals_no_sci_notation.csv')
df['amount_hex'] = df['amount'].astype(float).apply(lambda x: hex(int(x))).str.lstrip('0x')

df = df.drop(['is_unclaimed', 'amount'], axis=1)
df = df.rename(columns={'amount_hex': 'amount_tokens'})

df.loc[df['amount_tokens'] == '', 'amount_tokens'] = '0'

df.to_csv('csv_outputs/unclaimed_withdrawals_hex_output.csv', index=False)
display(df)
```
