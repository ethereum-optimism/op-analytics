```python
import pandas as pd
import requests as r

chains = [
     ['Zora Network','explorer.zora.energy']
    ,['Public Goods Network','explorer.publicgoods.network']
]
```

```python
def get_tx_from_blockscout(name, explorer_url, rm_system_txs = True, block_time = 2):
    #clean url
    explorer_url = explorer_url.replace('https://','')
    explorer_url = explorer_url.replace('http://','')
    #generate api string
    api_str = 'https://' + explorer_url + '/api/v2/stats/charts/transactions'
    res = r.get(api_str).json()['chart_data']
    df = pd.DataFrame(res)
    df['chain'] = name
    if rm_system_txs:
        df['tx_count'] = df['tx_count'] - (60*60*24)/(block_time) #2 second block time
    return df
```

```python
chaindf = []
for c in chains:
    chain_name = c[0]
    chain_url = c[1]
    tmp = get_tx_from_blockscout(chain_name, chain_url)
    chaindf.append(tmp)

chaindf = pd.concat(chaindf)
display(chaindf)

chaindf.to_csv('outputs/blockscout_api_txs' 
               + '_'
               + str(min(chaindf['date']))
               + '_'
               + str(max(chaindf['date']))
               + '.csv'
                )
```
